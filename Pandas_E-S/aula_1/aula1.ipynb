## **Parámetros de la función read_csv**


Documentación: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

Parámetro **sep** para definir tipo de separación en una base de datos que no está separada por comas

```python
datos_punto_coma = pd.read_csv("superstore_data_punto_coma.csv", sep = ";")
```

```python
datos_punto_coma.head()
```

Parámetro **nrows** que muestra la cantidad de líneas indicadas, en este caso, estará mostrando las 5 primeras líneas

```python
datos_primeras_lineas = pd.read_csv("superstore_data.csv", nrows=5)
datos_primeras_lineas
```

Parámetro **usecols** permite seleccionar una cantidad de columnas concretas y así solo contemplar datos de interés. En este caso se selecciona la columna de identificación, año de nacimiento y el ingreso que dio dicha persona a la tienda

```python
datos_seleccion = pd.read_csv("superstore_data.csv", usecols=["Id","Year_Birth","Income"])
datos_seleccion
```

```python
datos_seleccion_posicion = pd.read_csv("superstore_data.csv", usecols=[0,1,4])
datos_seleccion_posicion
```
